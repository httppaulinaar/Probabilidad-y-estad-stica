```python
from statistics import*
```

```python
#lista de calificaciones del grupo A de probabilidad y estadística
grupoa=[70,80,90,40,70,56,90,100,98]
print(grupoa)
type(grupoa)
```

```python
#la función type nis devuelve el tipo de dato de la variable
type(grupoa)
```

```python
#sacar la media de las calificaciones con una fórmula
promedio=sum(grupoa)/len(grupoa)
print(promedio)
```

```python
promedio2=mean(grupoa)
print(promedio2)
```

```python
#Determinar el valor de la mediana con la funcion median
mediana=median(grupoa)
print(mediana)
```

```python
#determinar el valor de la moda con la funcion mode
moda=mode(grupoa)
print(moda)
```

```python
#en caso de que exista más de una moda, se emplea la funcion multimode
moda2=multimode(grupoa)
print(moda2)
```

#ejemplo 2

```python
#definir los tiempos que tardaron los alumnospara correr los 100 metros
tiempos100mts=[15,20,23.1,41.5,12,19,16.3,14,10.5,10.3]
print(tiempos100mts)
```

```python
promediot=mean(tiempos100mts)
print("El tiempo promedio de los corredores es:",promediot)
```

```python
medianat=median(tiempos100mts)
print("La mediana de los tiempos de los corredores es: ",medianat)
```

```python
modat=mode(tiempos100mts)
print("La moda de los tiempos de los corredores es: ",modat)
```
