```python
from statistics import*
```

```python
#lista de calificaciones del grupo A de probabilidad y estadística
grupoa=[70,80,90,40,70,56,90,100,98]
print(grupoa)
type(grupoa)
```

```python
#la función type nis devuelve el tipo de dato de la variable
type(grupoa)
```

```python
#sacar la media de las calificaciones con una fórmula
promedio=sum(grupoa)/len(grupoa)
print(promedio)
```

```python
promedio2=mean(grupoa)
print(promedio2)
```

```python
#Determinar el valor de la mediana con la funcion median
mediana=median(grupoa)
print(mediana)
```

```python
#determinar el valor de la moda con la funcion mode
moda=mode(grupoa)
print(moda)
```

```python
#en caso de que exista más de una moda, se emplea la funcion multimode
moda2=multimode(grupoa)
print(moda2)
```

#Ejemplo 3

```python
#libreria para aleatorios
import random
```

```python
fabrica1=[]
for i in range(100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range(100):
  numero_aleatorio=round(random.uniform(70,100),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
#librería de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Producción diaria de la fábrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Producción diaria de la fábrica 2")
```

```python
print("El promedio de la fábrica 1 es: ",mean(fabrica1))
print("La moda de la fábrica 1 es: ",mode(fabrica1))
print("La mediana de la fábrica 1 es: ",median(fabrica1))
```

```python
print("El promedio de la fábrica 2 es: ",mean(fabrica2))
print("La moda de la fábrica 2 es: ",mode(fabrica2))
print("La mediana de la fábrica 2 es: ",median(fabrica2))
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```
