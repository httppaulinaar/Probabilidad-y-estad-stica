```python
import random

import matplotlib.pyplot as plt

from statistics import *
```

```python
fabrical = []

for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2)
    fabrical.append(numero_aleatorio)

print(fabrical)
```

```python
fabrica2 = []

for i in range(100):
    numero_aleatorio = round(random.uniform(48, 93), 2)
    fabrica2.append(numero_aleatorio)

print(fabrica2)
```

```python
plt.boxplot(fabrical)
plt.ylabel('Produccion en mts.')
plt.title('Produccion de la fabrica 1')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

ax1.boxplot(fabrical, vert=True, patch_artist=True)
ax1.set_title('Producción fabrica 1')

ax2.boxplot(fabrica2, vert=True, patch_artist=True)
ax2.set_title('Produccion fabrica 2')

plt.show()
```

```python
# Insertamos un valor de 150 a la fabrica 1.
# Sera un valor atípico
fabrical.append(150)
print(fabrical)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

ax1.boxplot(fabrical, vert=True, patch_artist=True)
ax1.set_title('Producción fabrica 1')

ax2.boxplot(fabrica2, vert=True, patch_artist=True)
ax2.set_title('Producción fabrica 2')

plt.show()
```

```python
# Agregamos ax1.set_ylim(30, 160) para garantizar que ambos gráficos usen la misma escala vertical
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 5))

ax1.boxplot(fabrical, vert=True, patch_artist=True)
ax1.set_title('Producción fabrica 1')
ax1.set_ylim(30, 160)

ax2.boxplot(fabrica2, vert=True, patch_artist=True)
ax2.set_title('Producción fabrica 2')
ax2.set_ylim(30, 160)

plt.show()
```
