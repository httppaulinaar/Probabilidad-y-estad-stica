## **Programa 6:  Estadística descriptiva**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
autos = pd.read_csv("autos2.csv")

autos.head()
```

```python

prices = autos['price'].dropna()

precio_min = prices.min()
precio_max = prices.max()
precio_moda = prices.mode()[0]
precio_media = prices.mean()
precio_mediana = prices.median()
precio_std = prices.std()

print("Valor mínimo del precio:", precio_min)
print("Valor máximo del precio:", precio_max)
print("Moda del precio:", precio_moda)
print("Media del precio:", precio_media)
print("Mediana del precio:", precio_mediana)
print("Desviación estándar del precio:", precio_std)
```

```python
autos[['price']].describe()
```

```python
# Regla de Sturges
n = prices.count()
bins_sturges = int(1 + 3.322 * np.log10(n))

print("Número óptimo de bins (Sturges):", bins_sturges)
```

```python
# Histograma
plt.figure()
plt.hist(prices, bins=bins_sturges)
plt.title("Histograma de precios de los automóviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure()
autos.boxplot(column='price', by='body-style')
plt.title("Precio según tipo de vehículo")
plt.suptitle("")
plt.xlabel("Tipo de vehículo")
plt.ylabel("Precio")
plt.show()
```

```python
plt.figure()
autos.boxplot(column='price', by='engine-location')
plt.title("Precio según ubicación del motor")
plt.suptitle("")
plt.xlabel("Ubicación del motor")
plt.ylabel("Precio")
plt.show()
```

# Conclusión general:
# El análisis estadístico muestra que el precio de los automóviles
# depende fuertemente del tipo de vehículo, la tracción y la ubicación del motor.
# La mayoría de los precios se concentran en rangos bajos y medios,
# mientras que los valores altos corresponden a configuraciones específicas.
